### last_login

Change data to datetime format.

```python
def strToDate(df,i=0):
    l = len(df)
    for dt in range(i,l):
        df[dt] = datetime.strptime(df[dt][0:10],"%Y-%m-%d").date()
    return df
```

```python
lLDate = userComb["last_login"]
lLDate = strToDate(lLDate)
```

```python
userComb["last_login"]=lLDate
userComb.head()
```
